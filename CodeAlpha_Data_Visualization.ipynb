The objective of this project is to transform raw sales data into meaningful visualizations that help identify business trends, customer behavior, regional performance, and profitability using Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
```

```python
# Load the dataset
df = pd.read_csv("SampleSuperstore.csv")

# Display first 5 rows
df.head()
```

### Observations

- The dataset contains sales transactions across multiple regions, categories, and customer segments.
- It includes both categorical and numerical features.
- Missing values and duplicate records were checked before visualization.
- The dataset is ready for creating business visualizations.

## 1. Sales by Category

This chart compares total sales across different product categories.

```python
sales_category = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))

plt.bar(sales_category.index, sales_category.values)

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

**Observation:**

- Technology generated the highest sales.
- Office Supplies contributed steady sales.
- Furniture generated comparatively lower sales.

## 2. Sales by Region

This chart shows sales performance across different regions.

```python
sales_region = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))

plt.bar(sales_region.index, sales_region.values)

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

**Observation:**

- The West region recorded the highest sales.
- The South region had comparatively lower sales.

## 3. Profit by Category

This visualization compares total profit across product categories.

```python
profit_category = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))

plt.bar(profit_category.index, profit_category.values)

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()
```

**Observation:**

- Technology generated the highest profit.
- Furniture had the lowest profit among all categories.

## 4. Quantity Sold by Category

This chart shows the total quantity sold for each category.

```python
quantity_category = df.groupby('Category')['Quantity'].sum()

plt.figure(figsize=(8,5))

plt.bar(quantity_category.index, quantity_category.values)

plt.title('Quantity Sold by Category')
plt.xlabel('Category')
plt.ylabel('Quantity')

plt.show()
```

**Observation:**

- Office Supplies had the highest quantity sold.
- Technology products were sold in lower quantities but generated higher revenue.

## 5. Ship Mode Distribution

This chart shows the number of orders shipped using different shipping modes.

```python
ship_mode = df['Ship Mode'].value_counts()

plt.figure(figsize=(8,5))

plt.bar(ship_mode.index, ship_mode.values)

plt.title('Ship Mode Distribution')
plt.xlabel('Ship Mode')
plt.ylabel('Number of Orders')

plt.xticks(rotation=15)

plt.show()
```

**Observation:**

- Standard Class is the most frequently used shipping mode.
- Same Day shipping is used the least.

## 6. Sales by Customer Segment

This visualization compares total sales across customer segments.

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(8,5))

plt.bar(segment_sales.index, segment_sales.values)

plt.title('Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Sales')

plt.show()
```

**Observation:**

- Consumer segment contributes the highest sales.
- Home Office segment contributes the lowest sales.

## 7. Discount vs Profit

This scatter plot shows the relationship between discount and profit.

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Discount'], df['Profit'])

plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')

plt.show()
```

**Observation:**

- Higher discounts often result in lower profits.
- Some high-discount transactions generate losses.

## 8. Correlation Heatmap

This heatmap displays relationships between numerical variables.

```python
numeric_df = df.select_dtypes(include=['number'])

corr = numeric_df.corr()

plt.figure(figsize=(8,6))

plt.imshow(corr, cmap='coolwarm', interpolation='nearest')

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title('Correlation Heatmap')

plt.show()
```

**Observation:**

- Sales and Profit have a positive relationship.
- Discount negatively affects Profit.
- Quantity has a weak correlation with Sales.

## 9. Top 10 States by Sales

This chart displays the top 10 states based on total sales.

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

plt.bar(top_states.index, top_states.values)

plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')

plt.xticks(rotation=45)

plt.show()
```

**Observation:**

- California generated the highest sales.
- The top 10 states contribute a significant portion of total sales.

## 10. Top 10 Sub-Categories by Profit

This visualization shows the most profitable product sub-categories.

```python
top_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

plt.bar(top_profit.index, top_profit.values)

plt.title('Top 10 Sub-Categories by Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')

plt.xticks(rotation=45)

plt.show()
```

**Observation:**

- Copiers and Phones are among the most profitable sub-categories.
- High-profit sub-categories contribute significantly to overall business profitability.
